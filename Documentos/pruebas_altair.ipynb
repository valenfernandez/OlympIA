```python
# Import necessary libraries
import pandas as pd
import altair as alt

# Sample data
data = {
    'archivo_origen': ['A', 'A', 'B', 'B', 'C', 'C'],
    'numero_linea': [1, 3, 5, 8, 12, 14],
    'clasificacion': ['V', 'NV', 'NV', 'V', 'NV', 'V'],
    'text': ['Text 1', 'Text 2', 'Text 3', 'Text 4', 'Text 5', 'Text 6']
}

df_violentos = pd.DataFrame(data)

# Domain and range for color scale
domain_v = ['V', 'NV']
range_ = ['red', 'blue']

scatterplots = []

for file_name, group_df in df_violentos.groupby('archivo_origen'):
    min_linea = group_df['numero_linea'].min()
    max_linea = group_df['numero_linea'].max()
    interval = max_linea - min_linea
    nice_value = max(1, int(interval / 10))

    x_scale = alt.Scale(nice=nice_value)

    scatterplot = alt.Chart(group_df, width=900).mark_circle().encode(
        x=alt.X('numero_linea:Q', title='Numero de lineas', scale=x_scale, bin=True),
        y=alt.Y('clasificacion:N', title='Categoria'),
        color=alt.Color('clasificacion:N', title='Categoria', scale=alt.Scale(domain=domain_v, range=range_)),
        tooltip=['clasificacion:N', 'count():Q', 'numero_linea:Q']
    ).properties(
        title=f'Relaci√≥n entre numero de linea y frases violentas en archivo {file_name}'
    )
    scatterplots.append(scatterplot)

chart_lineas_cat = alt.vconcat(*scatterplots)
chart_lineas_cat
```

```python
# Import necessary libraries
import pandas as pd
import altair as alt

# Sample data
data = {
    'archivo_origen': ['A', 'A', 'A', 'A', 'A', 'A'],
    'numero_linea': [1, 3, 5, 88, 123, 140],
    'clasificacion': ['V', 'NV', 'NV', 'V', 'NV', 'V'],
    'text': ['Text 1', 'Text 2', 'Text 3', 'Text 4', 'Text 5', 'Text 6']
}

df_violentos = pd.DataFrame(data)

# Domain and range for color scale
domain_v = ['V', 'NV']
range_ = ['red', 'blue']

scatterplots = []

# Define the desired number of ticks on the x-axis
num_ticks = 10

for file_name, group_df in df_violentos.groupby('archivo_origen'):
    max_v = group_df['numero_linea'].max()
    min_v = group_df['numero_linea'].min()
    range_v = max_v - min_v

    # Calculate the interval based on the range and desired number of ticks
    interval = max(1, int(range_v / num_ticks))

    vector = list(range(min_v, max_v + interval, interval))

    scatterplot = alt.Chart(group_df, width=900).mark_circle().encode(
        x=alt.X('numero_linea:Q', title='Numero de linea', axis=alt.Axis(values=vector)),
        y=alt.Y('clasificacion:N', title='Categoria'),
        color=alt.Color('clasificacion:N', title='Categoria', scale=alt.Scale(domain=domain_v, range=range_)),
        tooltip=['clasificacion:N', 'numero_linea:Q', 'text:N']
    ).properties(
        title=f'Relaci√≥n entre numero de linea y frases violentas en archivo {file_name}'
    )
    scatterplots.append(scatterplot)

chart_lineas_cat = alt.vconcat(*scatterplots)
chart_lineas_cat
```
