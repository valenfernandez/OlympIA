### Eliminacion de repetidos y espacios

```python
import openpyxl
import csv
import re

doc='cleaned_tweets.xlsx'
first_cell='A1'
last_cell= 'A1875'
phrases = []
count = 0
excel_document = openpyxl.load_workbook(doc, data_only=True)
first_sheet = excel_document.sheetnames[0]
sheet = excel_document[first_sheet]
for row in sheet[first_cell:last_cell]:
    for cell in row:
        if cell.value and cell.value != ' ' and cell.value :  # si la celda tiene informacion
            phrases.append(cell.value)
tweets_unicos = set(phrases)
with open('cleaned_tweets.csv', 'w', newline='') as output_file:
    writer = csv.writer(output_file)
    for tweet in tweets_unicos:
      writer.writerow([tweet])
      count = count + 1

print(count)
```
